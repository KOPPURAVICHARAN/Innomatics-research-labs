```python
import pandas as pd
import numpy as np
```

```python
unique_user_ids = movies1['userId'].nunique()
unique_user_ids
```

```python
df_movies = pd.read_csv('movies.csv')
df_movies
```

```python
movie_title = "Terminator 2: Judgment Day (1991)"
movie_id = df_movies[df_movies['title'] == movie_title]['movieId'].values[0]
movie_ratings = movies1[movies1['movieId'] == movie_id]
average_rating = movie_ratings['rating'].mean()
average_rating
```

```python
import matplotlib.pyplot as plt
movie_title = "Fight Club (1999)"
movie_id = df_movies[df_movies['title'] == movie_title]['movieId'].values[0]
movie_ratings = movies1[movies1['movieId'] == movie_id]
plt.figure(figsize=(10, 6))
plt.hist(movie_ratings['rating'], bins=10, edgecolor='black')
plt.title(f'Distribution of User Ratings for "{movie_title}"')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.xticks([0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0])
plt.grid(True)
plt.show()
```

```python
movie_rating_counts = movies1['movieId'].value_counts()
most_rated_movie_id = movie_rating_counts.idxmax()
most_rated_movie_count = movie_rating_counts.max()
most_rated_movie_title = df_movies[df_movies['movieId'] == most_rated_movie_id]['title'].values[0]
most_rated_movie_title
```

```python
grouped_ratings = movies1.groupby('movieId').agg(
    rating_count=('rating', 'count'),
    average_rating=('rating', 'mean')
).reset_index()
merged_df = pd.merge(df_movies, grouped_ratings, on='movieId')
popular_movies = merged_df[merged_df['rating_count'] > 50]
popular_movies
```

```python
most_popular_movie = popular_movies.loc[popular_movies['average_rating'].idxmax()]
most_popular_movie
```

```python
top_5_popular_movies = popular_movies.sort_values(by='rating_count', ascending=False).head(5)
top_5_popular_movies 
```

```python
sci_fi_movies = popular_movies[popular_movies['genres'].str.contains('Sci-Fi')]
third_most_popular_sci_fi = sci_fi_movies.sort_values(by='rating_count', ascending=False).iloc[2]
third_most_popular_sci_fi
```

```python
df_links = pd.read_csv("links.csv")
df_links
```

```python
merged_df = pd.merge(popular_movies, df_links, on='movieId')
merged_df['imdbId'] = merged_df['imdbId'].astype(str)
highest_imdb_id_movie = merged_df.loc[merged_df['imdbId'].idxmax()]
highest_imdb_id_movie
```

```python
sci_fi_movies = merged_df[merged_df['genres'].str.contains('Sci-Fi', case=False, na=False)]
sci_fi_movies['imdbId'] = sci_fi_movies['imdbId'].astype(str)
highest_imdb_id_sci_fi = sci_fi_movies.loc[sci_fi_movies['imdbId'].idxmax()]
highest_imdb_id_sci_fi
```
